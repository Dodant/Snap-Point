```python
y_44, sr = lisa.load('./sound/snap_44.wav')
mfcc = librosa.feature.mfcc(y=y_44, sr=sr)
print(y_44.shape)
print(mfcc.shape)
plt.figure(figsize=(14, 5))
lisa.display.waveplot(y_44, sr=sr)
plt.show()

# 3. Run the default beat tracker
tempo, beat_frames = librosa.beat.beat_track(y=y_44, sr=sr)

# 4. Convert the frame indices of beat events into timestamps
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
print(beat_times)

# lisa.display.specshow(mfcc, x_axis='time')
# plt.colorbar()
# plt.title('MFCC')
# plt.tight_layout()
# plt.show()

S = librosa.feature.melspectrogram(y_44, sr=sr, n_mels=128) 
 
log_S = librosa.amplitude_to_db(S)
plt.figure(figsize=(12, 4))
librosa.display.specshow(log_S, sr=sr, x_axis='time', y_axis='mel')
plt.title('mel power spectrogram')
plt.colorbar(format='%+02.0f dB')
plt.tight_layout()
plt.show()
```

```python
X = lisa.stft(y_44)
Xdb = lisa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
lisa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```
